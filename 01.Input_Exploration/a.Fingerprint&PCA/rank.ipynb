```python
import pandas as pd
import matplotlib.pyplot as plt

# 
data = [0.25488582, 0.10151231, 0.07920701, 0.07024532, 0.05530014, 0.0500565,
 0.04738598, 0.04175964, 0.03647973, 0.03317591]

import matplotlib.pyplot as plt

# 
data = [
    0.25488582,
    0.10151231,
    0.07920701,
    0.07024532,
    0.05530014,
    0.0500565,
    0.04738598,
    0.04175964,
    0.03647973,
    0.03317591,
]

# 
percentages = [f"{value * 100:.2f}%" for value in data]

# 
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(data)), data, color="skyblue")

# 
for i, bar in enumerate(bars):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        percentages[i],
        ha="center",
        va="bottom",
        fontsize=10,
    )

# 
plt.title("Explained Variance Ratio by Principal Components", fontsize=16)
plt.xlabel("Principal Components", fontsize=14)
plt.ylabel("Explained Variance Ratio", fontsize=14)

#  x 
plt.xticks(
    range(len(data)), [f"PC{i+1}" for i in range(len(data))], fontsize=10
)

# 
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# 
data = [0.10035947, 0.07978976, 0.06473562, 0.05427052, 0.05136736, 0.04590267,
 0.04059067, 0.03989571, 0.03738822, 0.03429235]

# 
percentages = [f"{value * 100:.2f}%" for value in data]

# 
colors = cm.viridis(np.linspace(0.2, 0.8, len(data)))

# 
plt.figure(figsize=(10, 8))
bars = plt.bar(range(len(data)), data, color=colors)

# 
for i, bar in enumerate(bars):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        percentages[i],
        ha="center",
        va="bottom",
        fontsize=12,
    )

# 
plt.title("Explained Variance Ratio of Morgan", fontsize=18)
plt.xlabel("Principal Components", fontsize=14)
plt.ylabel("Explained Variance Ratio", fontsize=14)

#  x 
plt.xticks(
    range(len(data)), [f"PC{i+1}" for i in range(len(data))], fontsize=14
)

# 
plt.tight_layout()
plt.show()
```
